```python
import torch
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def simulator(x, t=3):

    x0, v0 = x[:2], x[2:]

    a = torch.tensor([0, -9.81])
    v = v0 + a * t
    x = x0 + v0 * t + .5 * a * t**2

    return torch.cat((x, v))
```

```python
x0 = torch.tensor([0, 0])
v0 = torch.tensor([10, 10])

x = torch.cat((x0, v0))
```

```python
y = simulator(x, t=3)
```
